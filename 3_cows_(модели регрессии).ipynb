```python
# Создадим список столбцов, в которых будем искать неявные дубликаты
list_of_columns_to_find_duplicates = [data_ferma_main['breed'],
                                      data_ferma_main['pasture_type'],
                                      data_ferma_main['dad_breed'],
                                      data_ferma_dad['dad_name'],
                                      data_cow_buy['breed'],
                                      data_cow_buy['pasture_type'],
                                      data_cow_buy['dad_breed'],
                                      data_cow_buy['dad_name']]

# Создадим функцию, врозвращающую все уникальные значения столбца
def search_for_implicit_duplicates(list_of_columns):
    for column in list_of_columns:
        print(column.unique())
```

```python
# Вызывем функцию, врозвращающую все уникальные значения столбца
search_for_implicit_duplicates(list_of_columns_to_find_duplicates)
```

## Корреляционный анализ признаков ('ferma_main.csv')

### Изучение взаимосвязей между всеми признаками

Перед изучением взаимосвязи между признаками необходимо посмотреть какое распределение имеют количественные признаки (нормальное или ненормальное).

```python
# Создадим функцию, которая переберет список количественных признаков определит, какое распределение имеет каждый из них
col_sign = [data_ferma_main['milk_yield_kg'],
       data_ferma_main['energy_feed_unit'],
       data_ferma_main['crude_protein_g'],
       data_ferma_main['sugar_protein_ratio'],
       data_ferma_main['fat_content_%'],
       data_ferma_main['protein_%']]

def distribution_calculation(data):
    results = [st.chisquare(col) for col in data]
    p_values = [result[1] for result in results]
    is_normal = [p_value >= 0.05 for p_value in p_values]
    return is_normal

# Получаем список булевых значений характеристик распределения признаков и преобразуем его в датафрейм для создания таблицы с
# результатами
distribution_list = distribution_calculation(col_sign)
distribution_df = pd.DataFrame(distribution_list).rename(columns={0:'Признак'})
distribution_df['Количественный признак'] = ['milk_yield_kg',
                                             'energy_feed_unit',
                                             'crude_protein_g',
                                             'sugar_protein_ratio',
                                             'fat_content_%',
                                             'protein_%']
# На основании булевых значенй создаем столбец датафрейма, содержащий сведения о характеристиках распределения данных
def categorize_distribution(row):
    if row['Признак'] == False:
        return 'Данные распределены ненормально'
    else:
        return 'Данные распределены нормально'

distribution_df['Характеристика распределения данных'] = distribution_df.apply(categorize_distribution, axis=1)
distribution_df.drop(columns=['Признак']).set_index('Количественный признак')
```

Видим, что данные признаков 'milk_yield_kg' и 'crude_protein_g' распределены ненормально, данные же признаков 'energy_feed_unit', 'sugar_protein_ratio', 'fat_content_%' и 'protein_%' имеют нормальное распределение.

Теперь с целью изучения взаимосвязей между признаками построим матрицу корреляции количественных признаков датафрейма 'data_ferma_main'. по коэффициенту Спирмена.

## Обучение модели линейной регрессии

###  Первая модель LinearRegression

#### Подготовка данных и обучение на них первой модель LinearRegression (датасет 'ferma_main.csv')

В качестве целевого признака используем признак "Удои молока". В качестве входных признаков используем следующие признаки: "ЭКЕ", "Сырой протеин", "СПО", "Порода коровы", "Тип пастбища", "Порода папы быка" и "Возраст". Выбор указанных входных признаков обусловлен тем, что целевой признак на прямую связан с перечисленными признаками. Так, удойность коровы зависит как от ее породы и пастбища, на котором она паслась, так и порода ее папы и возраста коровы. Также на удойность влияют характеристиками коровьего корма ("ЭКЕ", "Сырой протеин" и "СПО"). При этом жирность молока, белок в молоке и вкус молока для нужд линейной регрессии нас не интересуют, так как являются показателями, описывающими само молока и не могут виять на удои молока.

```python
# Сохраним входные признаки в переменную X
X = data_ferma_main.drop(columns=['id', 'milk_yield_kg', 'fat_content_%', 'protein_%', 'milk_taste'])

# Сохраним целевой признак в переменную y
y = data_ferma_main['milk_yield_kg']

# Взглянем на первые пять строк переменных 'X' и 'y' для самопроверки
print(X.head())
y.head()
```

Входные и целевой показатели успешно сохранены в переменные 'X' и 'y' соответственно.\
Разобьем данные на тренировочную и тестовую выборки.

```python
RANDOM_STATE = 42

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    random_state=RANDOM_STATE
)
```

Кодируем категориальные признаки с помощью метода 'OneHotEncoder'.

```python
encoder = OneHotEncoder(drop='first', sparse_output=False)

cat_col_names = ['breed', 'pasture_type', 'dad_breed', 'age']

X_train_ohe = encoder.fit_transform(X_train[cat_col_names])

X_test_ohe = encoder.transform(X_test[cat_col_names])

encoder_col_names = encoder.get_feature_names_out()
```

Масштабируем количественные признаки с помощью метода 'StandardScaler'.

```python
scaler = StandardScaler()

num_col_names = ['energy_feed_unit', 'crude_protein_g', 'sugar_protein_ratio']

X_train_scaled = scaler.fit_transform(X_train[num_col_names])

X_test_scaled = scaler.transform(X_test[num_col_names])
```

Cоздадим тренировочный и тестовый датафреймы из закодированных и отмасштабированных данных.

```python
X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)

X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)
```

Обучим модель линейной регрессии на подготовленных данных.

```python
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)

# Получим предсказания модели на тестовых данных
predictions = model_lr.predict(X_test)
predictions
```

#### Оценка качества модели на тестовой выборке (R², анализ остатков)

```python
r2 = r2_score(y_test, predictions)
print("R2 LinearRegression =", r2)

residuals = y_test - predictions

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
axes[0].hist(x=residuals)
axes[0].set_title('Гистограмма распределения остатков')
axes[0].set_xlabel('Остатки')

axes[1].scatter(x=predictions, y=residuals)
axes[1].set_xlabel('Предсказания модели')
axes[1].set_ylabel('Остатки')
axes[1].set_title('Анализ дисперсии')
plt.show()
```

**Выводы** по первой модели LinearRegression: значение R² модели составляет ~0.78.\

### Модель LogisticRegression

#### Подготовка данных и обучение на них модели LogisticRegression (датасет 'ferma_main.csv')

**Главное для фермера - это вкусное молоко.** Соответсвенно для решения поставленной им задачи: создание модели, которая в том числе предсказывает какая корова будет давать вкусное молоко, наибольшее значение имеет метрика precision. Так как показатель данной метрики равный 1.0 исключит ситуацию, в которой модель будет ошибочно прогнозировать, что молоко отдельной коровы будет являться вкусным. При этом метрика recall для решения задачи не так важна, так как если модель ошибочно спрогнозирует, что корова дает невкусное молоко, хотя в действительности она дает вкусное молоко, это не скажется негативно на выборе фермера при покупке коров и качестве реализуемого фермером молока. Главное для фермера не приобрести по результатам прогноза модели корову, которая будет давать невуксное молоко. Соответственно для решения поставленных фермером задачи необходимо привести метрику precision к 1.0 и свести ложноположительную ошибку (первого рода) к 0.

*С указанной целью закодируем значения, которые принимает признак 'milk_taste' следующим образом "вкусно" = 1 и "не вкусно" = 0.*

```python
data_ferma_main['milk_taste'] = data_ferma_main['milk_taste'].replace(['вкусно', 'не вкусно'], [1, 0])
# Смотрим корректно ли создался закодированлись значения признака 'milk_taste' для самопроверки
data_ferma_main.head(3)
```

```python
# Сохраняем входные и целевые признаки в переменные, к входным признакам относим признаки, которые могут влиять на вкус молока,
# в качестве целевого признака выбираем "Вкус молока"
X_log = data_ferma_main.drop(columns=['id', 'milk_yield_kg', 'fat_content_%', 'protein_%', 'milk_taste', 'energy_feed_unit', 'sugar_protein_ratio'])
y_log = data_ferma_main['milk_taste']

X_log_train, X_log_test, y_log_train, y_log_test = train_test_split(
    X_log,
    y_log,
    random_state=RANDOM_STATE
)

cat_col_names = ['breed', 'pasture_type', 'dad_breed', 'age', 'sugar_protein_ratio_category', 'dad_name']

X_log_train_ohe = encoder.fit_transform(X_log_train[cat_col_names])
X_log_test_ohe = encoder.transform(X_log_test[cat_col_names])

encoder_col_names = encoder.get_feature_names_out()

num_col_names = ['crude_protein_g', 'energy_feed_unit_2']

X_log_train_scaled = scaler.fit_transform(X_log_train[num_col_names])
X_log_test_scaled = scaler.transform(X_log_test[num_col_names])

X_log_train_ohe = pd.DataFrame(X_log_train_ohe, columns=encoder_col_names)
X_log_test_ohe = pd.DataFrame(X_log_test_ohe, columns=encoder_col_names)
X_log_train_scaled = pd.DataFrame(X_log_train_scaled, columns=num_col_names)
X_log_test_scaled = pd.DataFrame(X_log_test_scaled, columns=num_col_names)

X_log_train = pd.concat([X_log_train_ohe, X_log_train_scaled], axis=1)
X_log_test = pd.concat([X_log_test_ohe, X_log_test_scaled], axis=1)

clf = LogisticRegression()
clf.fit(X_log_train, y_log_train)

# Получим предсказания модели на тестовых данных
predictions_log = clf.predict(X_log_test)
predictions_log
```

Значение recall говорит о том, что модель допускает ложноотрицательную ошибку (второго рода) почти в 14% случаев - то есть в 14% случаев "не отлавливает" коров, дающих вкусное молоко в . Ложноотрицательные или ложноположительные.
